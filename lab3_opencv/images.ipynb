```
#img = cv2.imread('/content/drive/My Drive/Colab Notebooks/input/AlainDelon.jpg', 1)
img = cv2.imread('AlainDelon.jpg', cv2.IMREAD_COLOR)
print(img.shape )
#cv2.imshow(img)
h, w = img.shape[:2]
cv2.putText(img, 'Alain', (20, h-50), cv2.FONT_HERSHEY_TRIPLEX , 2, (255, 255, 255) , 2, cv2.LINE_AA)

img = cv2.resize(img, (w/3, h/3), interpolation = cv2.INTER_CUBIC)
center = (w/6,h/6)
angle = 45
scale = 1
M = cv2.getRotationMatrix2D(center, angle, scale)
res = cv2.warpAffine(img, M, (w/3, h/3))
#print(res)
cv2_imshow(res)
```

```
#img = cv2.imread('/content/drive/My Drive/Colab Notebooks/input/AlainDelon.jpg', 1)
img = cv2.imread('AlainDelon.jpg', cv2.IMREAD_COLOR)
print(img.shape )
#cv2.imshow(img)
h, w = img.shape[:2]
img = cv2.resize(img, (w/3, h/3), interpolation = cv2.INTER_CUBIC)
h, w = img.shape[:2]
cv2.putText(img, 'Alain', (w/20,h-h/20), cv2.FONT_HERSHEY_TRIPLEX , 1, (255, 255, 255) , 2, cv2.LINE_AA)


print(img.shape )
points1 = np.float32([[50,50],[100, 50],[50, 100]])
points2 = np.float32([[60,55],[125, 60],[15, 80]])

M_trl = np.float32([[1,0,w/2+75],[0,1, 0]])
res = cv2.warpAffine(img,M_trl,(w*3, h))

M_trf = cv2.getAffineTransform(points1, points2)
res = cv2.warpAffine(res, M_trf, (w*3, h))

#print(h1, w2)
cv2_imshow(res)
```
